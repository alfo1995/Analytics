```python
import pandas as pd
train=pd.read_csv('/Users/alfonsodamelio/Downloads/Sant_train.csv',delimiter=',')
test=pd.read_csv('/Users/alfonsodamelio/Downloads/Sant_test.csv',delimiter=',')
train['label'] = 'train'
test['label'] = 'test'
```

```python
print(len(train))
print(len(test))
```

```python
concat=pd.concat([train,test])
```

```python
concat.columns[concat.isnull().any()].tolist()
```

```python
concat=concat.drop('ID',axis=1)
```

```python
concat=pd.get_dummies(concat,columns=['var36'],dummy_na=True)
```

```python
concat.head()
```

```python
# Split your data
train = concat[concat['label'] == 'train']
test = concat[concat['label'] == 'test']

# Drop your labels
train = train.drop('label', axis=1)
test = test.drop('label', axis=1)
```

```python
X=train.drop(['TARGET'],axis=1).values
y=train['TARGET'].values
```

```python
from sklearn.cross_validation import train_test_split, cross_val_score
from sklearn.feature_selection import SelectKBest, chi2, f_classif

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
#scale the X coloumns
sc_X = StandardScaler()
#for the training set we need to fit it, then scale it
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
from xgboost import XGBClassifier
classifier = XGBClassifier()
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
